```python
def fill_property_sub_type_na(data):
    # Fill NaN values in 'property_sub_type_en' using values from 'property_type_en'
    data['property_sub_type_en'] = data.apply(
        lambda row: row['property_type_en'] if pd.isna(row['property_sub_type_en']) else row['property_sub_type_en'],
        axis=1
    )
    
    return data


# Apply the function to the dataset
df = fill_property_sub_type_na(df)

df.head(20)
```

```python
def transform_rooms(value):
    if pd.isna(value):
        return 'Unknown'
    if value in ['Studio', 'Single Room']:
        return 1
    elif value in ['Office']:
        return value
    elif 'B/R' in value:
        return int(value.split()[0])
    else:
        return 'Others'

# Apply the transformation to the 'rooms_en' column
df['rooms_en'] = df['rooms_en'].apply(transform_rooms)

df.head(20)
```

```python
df.drop(columns=['property_type_en'], inplace=True)
df.head()
```

```python
df['nearest_metro_en'] = df['nearest_metro_en'].fillna('Unknown')
```

```python
import category_encoders as ce

# Assuming 'data' is your DataFrame

# List of categorical columns to encode
categorical_columns = ['property_usage_en', 'property_sub_type_en', 'area_name_en', 'nearest_metro_en', 'rooms_en', 'trans_group_en']

# Initialize the binary encoder
encoder = ce.BinaryEncoder(cols=categorical_columns)

# Fit and transform to produce binary encoded data
data = encoder.fit_transform(df)

# Display the head of the DataFrame to see some of the encoded features


# Display the new structure of the DataFram
data.head()
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'X' are your features and 'y' is the target variable
X_train, X_test, y_train, y_test = train_test_split(data.drop('actual_worth', axis=1), np.log(data['actual_worth']), test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor  # or RandomForestClassifier

# Initialize the model
model = RandomForestRegressor(random_state=42)

# Train the model
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
y_pred = model.predict(X_test)

# Calculate R
r2 = r2_score(y_test, y_pred)
print("R:", r2)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
